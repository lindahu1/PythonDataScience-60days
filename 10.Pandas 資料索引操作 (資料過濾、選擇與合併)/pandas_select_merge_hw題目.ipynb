題目:<br>
讀取STOCK_DAY_0050_202009.csv串聯STOCK_DAY_0050_202010.csv，並且找出open小於close的資料


```python
import pandas as pd
```

```python
#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv
data1=pd.read_csv('STOCK_DAY_0050_202009.csv').set_index('date')
data2=pd.read_csv('STOCK_DAY_0050_202010.csv').set_index('date')
```

```python
data1
```

```python
#串聯
data=pd.concat([data1,data2])
```

```python
#找出open小於close的資料
```

```python
data.loc[(data.open<data.close),["open","close"]]
```
